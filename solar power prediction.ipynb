```python
import pandas as pd

# Load your dataset, assuming you have a CSV file or any other data source
df = pd.read_csv('solarpowergeneration.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

df[df.columns[:9]].hist(bins = 30, figsize=(15,10))
plt.show()
```

```python
df[df.columns[18:21]].hist(bins=30,figsize=(15,10))
plt.show()
```

```python
# Bivariate analysis
#scatter plot with target feature
plt.figure(figsize=(15,30))
for i,column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    plt.scatter(df[column],df['generated_power_kw'])
    plt.title(f'{column}vs Generated Power(kw)')
    plt.xlabel(column)
    plt.ylabel('Generated Power(kw)')
plt.tight_layout()
plt.show()
```
